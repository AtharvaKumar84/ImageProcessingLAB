```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Step 1: Read input image 
img = Image.open("lena.png").convert("L")  # L = grayscale
img = np.array(img)

# Step 2: Brightness Improvement
bright_up = np.clip(img + 50, 0, 255)

# Step 3: Brightness Reduction
bright_down = np.clip(img - 50, 0, 255)

# Step 4: Thresholding
thresh = np.where(img >= 128, 255, 0)

# Step 5: Negative Transformation
negative = 255 - img

# Step 6: Logarithmic Transformation
c = 255 / np.log(1 + np.max(img))
log_trans = c * np.log(1 + img)
log_trans = np.array(log_trans, dtype=np.uint8)

# Step 7: Power-law (Gamma) Transformation
gamma = 0.5
gamma_trans = np.array(255 * ((img / 255) ** gamma), dtype=np.uint8)

# Step 8: Display results
plt.figure(figsize=(12, 8))

plt.subplot(2, 4, 1); plt.imshow(img, cmap='gray'); plt.title("Original"); plt.axis("off")
plt.subplot(2, 4, 2); plt.imshow(bright_up, cmap='gray'); plt.title("Brightness +"); plt.axis("off")
plt.subplot(2, 4, 3); plt.imshow(bright_down, cmap='gray'); plt.title("Brightness -"); plt.axis("off")
plt.subplot(2, 4, 4); plt.imshow(thresh, cmap='gray'); plt.title("Thresholding"); plt.axis("off")
plt.subplot(2, 4, 5); plt.imshow(negative, cmap='gray'); plt.title("Negative"); plt.axis("off")
plt.subplot(2, 4, 6); plt.imshow(log_trans, cmap='gray'); plt.title("Log Transform"); plt.axis("off")
plt.subplot(2, 4, 7); plt.imshow(gamma_trans, cmap='gray'); plt.title(f"Gamma ={gamma}"); plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from PIL import Image

# Load lena.png in grayscale
img = Image.open("lena.png").convert("L")
img = np.array(img)

# Select a 5x5 patch (top-left corner)
sample_patch = img[0:5, 0:5]
print("Original Pixel Values (5x5 patch):\n", sample_patch)
```

```python

#Step 2  Brightness Improvement
bright_up = np.clip(img + 50, 0, 255)
print("\nBrightness Improved (5x5 patch):\n", bright_up[0:5, 0:5])
```

```python
#Step 3  Brightness Reduction

bright_down = np.clip(img - 50, 0, 255)
print("\nBrightness Reduced (5x5 patch):\n", bright_down[0:5, 0:5])
```

```python
#Step 4 - Thresholding
thresh = np.where(img >= 128, 255, 0)
print("\nThresholded (5x5 patch):\n", thresh[0:5, 0:5])
```

```python
#Step 5 - Negative Transformation

negative = 255 - img
print("\nNegative (5x5 patch):\n", negative[0:5, 0:5])
```

```python
#Step 6 - Logarithmic Transformation

c = 255 / np.log(1 + np.max(img))   
log_trans = c * np.log(1 + img)
log_trans = np.array(log_trans, dtype=np.uint8)
print("\nLog Transform (5x5 patch):\n", log_trans[0:5, 0:5])
```

```python
#Step 7 - Power Gamma Transformation

gamma = 0.5   # try 0.5 or 1.5
gamma_trans = np.array(255 * ((img / 255) ** gamma), dtype=np.uint8)
print(f"\nGamma Transform (={gamma}) (5x5 patch):\n", gamma_trans[0:5, 0:5])
```
